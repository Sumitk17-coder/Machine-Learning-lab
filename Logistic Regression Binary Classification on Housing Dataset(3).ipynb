```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder # to handle categorical features
```

```python

# load the dataset
data = pd.read_csv('/content/Housing.csv')
```

```python

# check the data to understand its structure
print("Dataset Head:")
print(data.head(10))
```

```python

# define the threshold price to categorize 'expensive' and 'affordable'
price_threshold = data['price'].median()
data['price_category'] = data['price'].apply(lambda x: 1 if x > price_threshold else 0)
```

```python

# prepare feeatures and target variable
x = data.drop(['price', 'price_category'], axis = 1)
y = data['price_category']
```

```python

# convert categorical features to numerical using Label Encoding
for column in x.select_dtypes(include=['object']).columns:
    label_encoder = LabelEncoder()
    x[column] = label_encoder.fit_transform(x[column])
```

```python

print(data.head(10))
```

```python

#split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

```

```python

# initialize and train the logistic regresson model
model = LogisticRegression(max_iter = 1000)
model.fit(X_train, y_train)
```

```python

# make predictions on the test set
y_pred = model.predict(X_test)
```

```python

# print the correct and incorrect predicitons
print ("/nPredictions with their correctness: ")
for actual, predicted in zip(y_test, y_pred):
  print(f"Actual: {actual}, Predicted: {predicted}, Correct: {actual == predicted}")
```

```python

# calculate and print the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("\nModel Accuracy:", accuracy)
```

```python

# Display confusion matrix for more insight
conf_matrix = confusion_matrix(y_test, y_pred)
print("/nConfusion Matrix:")
print(conf_matrix)
```
