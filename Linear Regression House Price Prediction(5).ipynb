```python
# Extract features and target
X = df[['var_1', 'var_2', 'var_3', 'var_4', 'var_5']]
y = df['output']
```

```python
🧩 1️⃣ Extracting Features
X = df[['var_1', 'var_2', 'var_3', 'var_4', 'var_5']]


✅ Purpose:

You are selecting specific columns from your DataFrame df that will be used as input variables (independent variables).

These columns — var_1, var_2, var_3, var_4, var_5 — are the features your model will learn patterns from.

🧠 Think of it like this:

The model will look at var_1 to var_5 to try to predict something (the target).

📦 Result:
X becomes a DataFrame like this:

var_1	var_2	var_3	var_4	var_5
12.4	34.1	8.2	10.1	5.4
13.5	31.9	9.0	11.2	6.1
...	...	...	...	...
```

```python
🎯 2️⃣ Extracting Target (Output)
y = df['output']


✅ Purpose:

Selects the target column, which contains the value you want the model to predict.

This is your dependent variable.

🧠 Think of it like this:

“Given the values of var_1–var_5, predict what output will be.”

📦 Result:
y becomes a Series (a single column) like:

output
45.2
50.1
...
```

```python
⚖️ Why Separate X and y?

Because most machine learning models in scikit-learn (like LinearRegression) expect data in this format:

model.fit(X, y)


X → input features (matrix form)

y → labels or target values (vector form)

🧠 Summary
Variable	Type	Description
X	DataFrame	Input features (independent variables)
y	Series	Target variable (dependent variable)
```

```python
# Split data into train-test sets (70-30 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
🧩 Code Explanation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


This line uses train_test_split() from sklearn.model_selection to divide your dataset into training and testing parts.

⚙️ Purpose

We split data to:

Train the model on one portion of the data (the training set)

Test (or evaluate) how well it performs on unseen data (the testing set)

This helps prevent overfitting (when a model memorizes the data instead of learning general patterns).

🧠 Parameter-by-Parameter Explanation
Parameter	Meaning	Example
X, y	Input features and target variable	Data you prepared earlier
test_size=0.3	30% of data is used for testing, 70% for training	70-30 split
random_state=42	Ensures the same random split every time (reproducible results)	If you change this number, you’ll get a different random split
📦 What You Get After Split
Variable	Description
X_train	Training features (used to train the model)
X_test	Testing features (used to evaluate performance)
y_train	Training labels (true outputs for X_train)
y_test	Testing labels (true outputs for X_test)
💡 Example Visualization
Dataset Portion	% of Data	Used For
Training Set (X_train, y_train)	70%	Model learns patterns
Testing Set (X_test, y_test)	30%	Model performance check
🧠 Behind the Scenes

The function randomly shuffles your data before splitting.

It ensures both sets represent the overall data distribution.

🧪 Mini Example

If your dataset has 100 rows:

70 rows → training

30 rows → testing
```

```python

# Create the pipeline with preprocessing and linear regression
pipeline = Pipeline([
    ('scaler', StandardScaler()),         # Step 1: Scale the features
    ('linear_regression', LinearRegression())  # Step 2: Train Linear Regression
])
```

```python
🎯 What is a Pipeline?

A Pipeline is like a machine learning assembly line.
Instead of you doing each step separately (scaling → training → predicting),
the pipeline connects all steps together in one object.

🧱 Breaking It Down
Step	Name	What It Does
1️⃣ ('scaler', StandardScaler())	“scaler”	Standardizes (normalizes) your features so all columns have similar ranges — mean = 0, standard deviation = 1.
2️⃣ ('linear_regression', LinearRegression())	“linear_regression”	Fits a Linear Regression model to the scaled data.
🧠 Why Use a Pipeline?

Without a pipeline, you would do this manually:

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

model = LinearRegression()
model.fit(X_train_scaled, y_train)


But with a pipeline, you can simply write:

pipeline.fit(X_train, y_train)


and it will automatically:

Scale the data

Train the regression model

Keep the exact same steps ready for prediction later

⚙️ In simple words:

A Pipeline = Combine multiple steps (like scaling + modeling) into one clean process.

So when you later do:

pipeline.predict(X_test)


It will:

Automatically scale your test data (same way as training data)

Make predictions using the trained Linear Regression model

💡 Analogy:

Imagine you run a factory 🏭 that makes juice:

Step 1: Wash the fruits 🍎

Step 2: Extract the juice 🧃

A Pipeline is like a machine that does both steps together automatically —
you just give it fruits (your data), and it gives you juice (predictions).
```

```python

# Fit the pipeline to the training data
model = pipeline.fit(X_train, y_train)
```

```python
⚙️ What Happens Here

You are training (or fitting) your pipeline on the training data.
Let’s unpack what happens inside the pipeline automatically 👇

🔁 Step-by-Step Process Inside the Pipeline
🧮 Step 1: StandardScaler() runs first

It looks at all the training features (X_train)

Calculates the mean and standard deviation for each column

Then scales all values so:

mean = 0

standard deviation = 1

👉 This makes sure no feature dominates because of a larger scale (like “Salary” vs “Age”).

Example before scaling:

var_1	var_2
50000	25
60000	30

After scaling:

var_1	var_2
-0.75	-1.1
0.75	1.1
📈 Step 2: LinearRegression() runs next

It takes the scaled data from step 1

Then learns a mathematical relationship between X (inputs) and y (output).

This is called training the model.

In simple words:

The model finds the best line (or equation) that fits the data points.

It learns something like:

output=𝑏0+𝑏1×𝑣𝑎𝑟1+𝑏2×𝑣𝑎𝑟2+...+𝑏5×𝑣𝑎𝑟5

Where:

𝑏0 = intercept (starting value)

𝑏1,𝑏2,...,... = coefficients (how much each variable affects the output)




🧠 Why model = pipeline.fit(...) ?

pipeline.fit() trains the model

It returns a trained pipeline object, which you store in model

Now, model is your trained machine learning model — ready to make predictions!

🪄 In plain English:

You just taught your model how to predict the output based on input data.
It learned patterns from your training dataset (X_train, y_train).
```

```python

# Make Predictions
y_pred = model.predict(X_test)
```

```python

# Evaluate the Model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R²: {r2}")
```

```python

# Residual Analysis
residuals = y_test - y_pred
```

```python

# Plot histogram of residuals
plt.figure(figsize=(5, 4))
sns.histplot(residuals, kde=True, bins=30, color='green')
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

```python

# Scatter Plot: Actual vs. Predicted
plt.figure(figsize=(5, 4))
plt.scatter(y_test, y_pred, s=15, c='green', alpha=0.7)
plt.plot(y_test, y_test, color='red', label="Ideal Fit Line")
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted')
plt.legend()
plt.show()
```

```python

# Extract Coefficients and Intercept from the Model
lr = model.named_steps['linear_regression']
print(f"Intercept: {lr.intercept_}")
print(f"Coefficients: {lr.coef_}")
```
